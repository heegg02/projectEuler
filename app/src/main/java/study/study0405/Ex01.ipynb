https://en.wikipedia.org/wiki/HTML 위의 사이트에서 각 종 태그를 불러오는 예제입니다~

```python
import urllib.request as req
from bs4 import BeautifulSoup as bs

html = req.urlopen('https://en.wikipedia.org/wiki/HTML')
soup = bs(html, 'html.parser')
```

```python
# 1
spans = soup.find_all('span', {'class':'vector-dropdown-label-text'})
for span in spans:
    print(span.string)
```

```python
# 2
spans = soup.find_all('span', {'class':'vector-dropdown-label-text'})
for span in spans:
    if 'b' in span.string:
        print(span.string)
```

```python
# 3
spans = soup.find_all('span', {'class':'vector-dropdown-label-text'})
for span in spans:
    if 'b' not in span.string:
        print(span.string)
```

```python
# 4
for tag in soup.find_all(class_=True):
    classes = tag.get('class')
    for class_ in classes:
        if 'vector' in class_:
            print(class_)
```

```python
# 5
for tag in soup.find_all(class_=True):
    classes = tag.get('class')
    for class_ in classes:
        if 'vector' in class_:
            if 15 >= len(class_):
                print(class_)
```
